# Text Splitting

```python
import tiktoken

# GPT 4.0 tokening
enc = tiktoken.get_encoding("cl100k_base")

def token_length(text):
    return len(enc.encode(text))
```

```python
!pip install langchain-text-splitters
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators = ['\n\n\n', '\n\n', '\n'],
    length_function = token_length,
    is_separator_regex = False,
    keep_separator = False,
    add_start_index = True,
    strip_whitespace = True
)

chunks = text_splitter.split_documents([doc])
```

# Vector Store

```python
!pip install langchain-community langchain-huggingface
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
!pip install chromadb
vector_store = Chroma(
    embedding_function = HuggingFaceEmbeddings(),
    ## GIVE A PERMANENT FOLDER NAME TO STORE VECTORS(DATABASE NAME)
    persist_directory = "CSJMU QUEST",
    ## GIVE A NAME TO YOUR COLLECTION(TABLE)
    collection_name = "broucher1"
)
```

```python
vector_store.add_documents(chunks)
```

```python
retriever = vector_store.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
```

```python
query = "admission process"

docs = retriever.invoke(query)

context = "\n\n".join([d.page_content for d in docs])
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="google/flan-t5-base"
)
```

```python
prompt = f"""
Answer the question using ONLY the given context.

Context:
{context}

Question:
{query}

Answer:
"""

response = pipe(prompt, max_length=300)

print(response[0]['generated_text'])
```
